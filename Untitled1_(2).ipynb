```python
def fourier_encode(x, num_encodings=4):
    encodings = [x]
    for i in range(num_encodings):
        for fn in [np.sin, np.cos]:
            encodings.append(fn((2.0 ** i) * x))
    return np.concatenate(encodings, axis=-1)
```

```python
def kl_anneal(step, cycle=1000, ratio=0.5):
    phase = step % cycle
    if phase < ratio * cycle:
        return phase / (ratio * cycle)
    else:
        return 1.0
```

```python
from torch_geometric.nn import GATv2Conv
from torch.nn import Linear, Dropout, BatchNorm1d
import torch.nn.init as init

from torch.nn import Linear, Dropout, BatchNorm1d
import torch.nn.init as init
from torch_geometric.nn import GATv2Conv

class EnhancedGATEncoder(torch.nn.Module):
    def __init__(self, in_channels, out_channels, edge_dim):
        super(EnhancedGATEncoder, self).__init__()
        self.dropout = Dropout(p=0.3)

        # Positional embedding (for x, y coords)
        self.pos_proj = Linear(2, 16)
        self.input_proj = Linear(in_channels, in_channels)  # Optional, but helps for stability

        # Edge feature projection
        self.edge_proj = Linear(edge_dim, 16)

        # GATv2 layers with increased input due to positional info
        self.gat1 = GATv2Conv(in_channels + 16, out_channels, heads=4, dropout=0.3, edge_dim=16)
        self.bn1 = BatchNorm1d(out_channels * 4)

        self.gat2 = GATv2Conv(out_channels * 4, out_channels, heads=2, dropout=0.3, edge_dim=16)
        self.bn2 = BatchNorm1d(out_channels * 2)

        self.mu_layer = GATv2Conv(out_channels * 2, out_channels, heads=1, concat=False, edge_dim=16)
        self.logvar_layer = GATv2Conv(out_channels * 2, out_channels, heads=1, concat=False, edge_dim=16)

        self.apply(self.init_weights)

    def init_weights(self, m):
        if isinstance(m, torch.nn.Linear) or isinstance(m, torch.nn.Conv1d):
            init.xavier_uniform_(m.weight)
            if m.bias is not None:
                init.zeros_(m.bias)

    def forward(self, x, edge_index, edge_attr, pos):
        # Project and concatenate positional features
        pos_emb = self.pos_proj(pos)
        x = self.input_proj(x)
        x = torch.cat([x, pos_emb], dim=1)

        edge_attr = self.edge_proj(edge_attr)

        x = self.dropout(F.elu(self.bn1(self.gat1(x, edge_index, edge_attr))))
        x = self.dropout(F.elu(self.bn2(self.gat2(x, edge_index, edge_attr))))

        mu = self.mu_layer(x, edge_index, edge_attr)
        logvar = self.logvar_layer(x, edge_index, edge_attr)
        return mu, logvar
```

```python
import torch.nn as nn
# class EdgeDecoder(torch.nn.Module):
#     def __init__(self, emb_dim):
#         super().__init__()
#         self.mlp = torch.nn.Sequential(
#             Linear(2 * emb_dim, 64),
#             torch.nn.ReLU(),
#             Linear(64, 1)
#         )

#     def forward(self, z, edge_index):
#         src = z[edge_index[0]]
#         dst = z[edge_index[1]]

#         if src.dim() == 1:
#             src = src.unsqueeze(0)
#         if dst.dim() == 1:
#             dst = dst.unsqueeze(0)

#         edge_input = torch.cat([src, dst], dim=1)
#         return self.mlp(edge_input).view(-1)
class EdgeDecoder(nn.Module):
    def __init__(self, latent_dim):
        super().__init__()
        self.mlp = nn.Sequential(
            nn.Linear(2 * latent_dim, 128),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(128, 6),
            nn.ReLU(),
            nn.Linear(6,1)
        )

    def forward(self, z, edge_index):
        src = z[edge_index[0]]
        dst = z[edge_index[1]]

        if src.dim() == 1:
            src = src.unsqueeze(0)
            dst = dst.unsqueeze(0)

        edge_input = torch.cat([src, dst], dim=1)
        return self.mlp(edge_input).view(-1)
```

```python
from torch_geometric.nn import VGAE, GCNConv
import torch.nn.functional as F
model = VGAE(EnhancedGATEncoder(in_channels, out_channels, edge_attr.shape[1])).to(device)
```

```python
encoder = EnhancedGATEncoder(in_channels, out_channels, edge_attr.shape[1]).to(device)
mu, logvar = encoder(train_data.x, train_data.edge_index, train_data.edge_attr, train_data.pos)
latent_dim = mu.shape[1]
decoder = EdgeDecoder(latent_dim=latent_dim).to(device)
optimizer = torch.optim.Adam(
        list(model.parameters()) + list(decoder.parameters()),
        lr=0.005,
        weight_decay=1e-4)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.5)
current_step = 0
```

```python
from torch_geometric.utils import negative_sampling

@torch.no_grad()
def test(model, decoder, data):
    model.eval()
    data = data.to(device)

    # Get embeddings from the model
    mu, logvar = encoder(data.x, data.edge_index, data.edge_attr, data.pos)
    logvar = torch.clamp(logvar, min=-10, max=10)  # Optional: helps with stability
    std = torch.exp(0.5 * logvar)
    eps = torch.randn_like(std)
    z = mu + eps * std
    # Sample latent z using reparameterization trick
    std = torch.exp(0.5 * logvar)
    eps = torch.randn_like(std)
    z = mu + eps * std  # Reparameterized sample

    # Positive edges
    pos_edge_index = data.edge_label_index
    pos_edge_label = data.edge_label

    # Negative sampling
    neg_edge_index = negative_sampling(
        edge_index=data.edge_index,
        num_nodes=z.size(0),
        num_neg_samples=pos_edge_index.size(1)
    )

    # Labels
    pos_y = torch.ones(pos_edge_index.size(1), device=z.device)
    neg_y = torch.zeros(neg_edge_index.size(1), device=z.device)
    y = torch.cat([pos_y, neg_y], dim=0)

    # Predict scores using the decoder
    pos_score = decoder(z, pos_edge_index)
    neg_score = decoder(z, neg_edge_index)
    scores = torch.cat([pos_score, neg_score], dim=0)

    # Apply sigmoid and compute metrics
    sigmoid_scores = torch.sigmoid(scores)

    auc = compute_auc(sigmoid_scores, y)
    ap = compute_ap(sigmoid_scores, y)

    return auc, ap


def compute_auc(scores, labels):
    from sklearn.metrics import roc_auc_score
    return roc_auc_score(labels.cpu().numpy(), scores.cpu().numpy())

def compute_ap(scores, labels):
    from sklearn.metrics import average_precision_score
    return average_precision_score(labels.cpu().numpy(), scores.cpu().numpy())
```

```python
for epoch in range(5000):
    model.train()
    optimizer.zero_grad()

    mu, logvar = encoder(train_data.x, train_data.edge_index, train_data.edge_attr, train_data.pos)
    logvar = torch.clamp(logvar, min=-10, max=10)  # Optional: helps with stability
    std = torch.exp(0.5 * logvar)
    eps = torch.randn_like(std)
    z = mu + eps * std
    # logvar = torch.clamp(logvar, min=-10, max=10)  # Prevent NaNs
    # std = torch.exp(0.5 * logvar)
    # eps = torch.randn_like(std)
    # z = mu + eps * std
    # print(z)
    # Compute the reconstruction loss
    edge_label_index = train_data.edge_label_index.to(dtype=torch.long, device=z.device)
    edge_label = train_data.edge_label.to(dtype=torch.float, device=z.device)
    logits = decoder(z, edge_label_index)
    recon_loss = F.binary_cross_entropy_with_logits(logits, edge_label)

    # recon_loss = torch.nn.functional.binary_cross_entropy_with_logits(
    #     (z[edge_label_index[0]] * z[edge_label_index[1]]).sum(dim=1), edge_label
    # )

    # KL divergence loss
    kl_weight = kl_anneal(current_step, cycle=1000)
    current_step += 1
    kl_loss = -0.5 * torch.mean(1 + logvar - mu.pow(2) - logvar.exp())
    loss = recon_loss + kl_weight * kl_loss
    # loss = recon_loss + kl_weight * model.kl_loss() / train_data.num_nodes

    loss.backward(retain_graph=True)

    # Clip gradients
    torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

    optimizer.step()
    scheduler.step()  # Step the scheduler

    if epoch % 5 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")

        val_data = val_data.to(device)
        auc, ap = test(model, decoder, val_data)
        print(f"Validation AUC: {auc:.4f}, AP: {ap:.4f}")
        print(f"Epoch {epoch}, Total Loss: {loss.item():.4f}, Recon: {recon_loss.item():.4f}, KL: {kl_loss.item():.4f} (weight: {kl_weight:.4f})")

```

```python
# Map OSM node IDs to contiguous indices
node_mapping = {node: i for i, node in enumerate(G.nodes())}

# Centrality measures and other network-based features
betweenness = nx.betweenness_centrality(G)
closeness = nx.closeness_centrality(G)
pagerank = nx.pagerank(G)

# Optional: Eccentricity measure (can be slow)
try:
    eccentricity = nx.eccentricity(G)
except:
    eccentricity = {node: 0 for node in G.nodes()}

# Prepare features
node_features = []
positions = []

for node, data in G.nodes(data=True):
    x_coord = data['x']
    y_coord = data['y']
    degree = G.degree[node]
    street_count = data.get('street_count', 0)
    bcent = betweenness[node]
    ccent = closeness[node]
    pr = pagerank[node]
    ecc = eccentricity[node]

    # New binary feature: is this node likely to have a traffic light?
    traffic_light = 1.0 if street_count == 4 else 0.0

    node_features.append([
        x_coord, y_coord, degree, street_count,
        bcent, ccent, pr, ecc,  # Removed clustering
        traffic_light
    ])
    positions.append([x_coord, y_coord])

# Convert to PyTorch tensors
x = torch.tensor(node_features, dtype=torch.float)
pos = torch.tensor(positions, dtype=torch.float)
```

```python
import pandas as pd
from google.colab import files

# Map OSM node IDs to contiguous indices
node_mapping = {node: i for i, node in enumerate(G.nodes())}

# Prepare edge features
edge_index = []
edge_features = []

highway_type_id = {
    'motorway': 0, 'trunk': 1, 'primary': 2, 'secondary': 3,
    'tertiary': 4, 'residential': 5, 'service': 6, 'unclassified': 7
}

for u, v, data in G.edges(data=True):
    u_id = node_mapping[u]
    v_id = node_mapping[v]

    length = data.get('length', 0.0)
    highway = data.get('highway', 'unclassified')
    if isinstance(highway, list):
        highway = highway[0]
    type_id = highway_type_id.get(highway, 7)

    maxspeed = data.get('maxspeed', '0')
    if isinstance(maxspeed, list):
        maxspeed = maxspeed[0]
    try:
        maxspeed = float(str(maxspeed).split()[0])
    except:
        maxspeed = 0.0

    lanes = data.get('lanes', '1')
    try:
        lanes = float(str(lanes).split()[0])
    except:
        lanes = 1.0

    bridge = 1.0 if data.get('bridge') in ['yes', 'true'] else 0.0
    tunnel = 1.0 if data.get('tunnel') in ['yes', 'true'] else 0.0
    oneway = 1.0 if data.get('oneway') in [True, 'yes', 'true', '1'] else 0.0

    edge_index.append([u_id, v_id])
    edge_features.append([
        u_id, v_id, length, type_id, maxspeed, lanes,
        bridge, tunnel, oneway
    ])

# Save edge features to CSV
edge_columns = [
    'source', 'target', 'length', 'road_type_id', 'maxspeed',
    'lanes', 'bridge', 'tunnel', 'oneway'
]
edge_df = pd.DataFrame(edge_features, columns=edge_columns)
edge_df.to_csv("edge_features_kasganj.csv", index=False)
# files.download("edge_features_kasganj.csv")
```
